## Import Data

```python
# import pandas and numpy
import numpy as np
import pandas as pd
import warnings
```

This data was gathered from participants in experimental speed dating events from 2002-2004. During the events, the attendees would have a four-minute "first date" with every other participant of the opposite sex. At the end of their four minutes, participants were asked if they would like to see their date again. They were also asked to rate their date on six attributes: Attractiveness, Sincerity, Intelligence, Fun, Ambition, and Shared Interests. The dataset also includes questionnaire data gathered from participants at different points in the process. These fields include: demographics, dating habits, self-perception across key attributes, beliefs on what others find valuable in a mate, and lifestyle information.

```python
# import dating data set
df = pd.read_csv("speeddating.csv")
df.head()
```

## Feature Engineering

For all of the fields with a float data type we are filling the "n/a"s with the median value for that field

```python
for col in df.columns:
    if df[col].dtype == "float64":
        # fill NA with median
        df[col] = df[col].fillna(df[col].median()) 
```

```python
# drop has null field
df.drop("has_null" , axis = 1, inplace = True)
df.head()
```

Next I want to encode all of the categorical fields. I created a list of categorical variables and passed these into the get dummies function as an argument

```python
encode_cols = list(df.select_dtypes(include = 'object'))

df2 = pd.get_dummies(df, columns = encode_cols, drop_first = True)
df2.head()
```

```python
df = df2

df.shape
```

```python
df.head()
```

The categorical fields all appear to be at the end of the data frame, therefore I will check for the position of the first categorical field and then change the data type for all of the following fields to numeric.

```python
# check column index of first categorical field
df.iloc[:,59]
```

```python
df.iloc[:,58]
```

```python
df.iloc[:, 59:].apply(pd.to_numeric, errors='coerce')
df.head()
```

```python
# check updated dtypes
df.iloc[:,59:].dtypes
```

```python
warnings.filterwarnings('ignore')
df.iloc[:,59:] = df.iloc[:,59:].astype(int)
```

```python
df.head()
```

I now want to separate my target variable from my feature variables. I drop some of the columns from the list of features because they give away too much information about the value of the target

```python
X = df.drop(["match_b'1'", 
             "decision_b'1'", 
             "decision_o_b'1'", 
             "d_like_b'[6-8]'", 
             "d_like_b'[9-10]'",
             "like"
            ], axis=1)


y = df["match_b'1'"]
```

```python
X.shape
```

```python
y.shape
```

```python
y.head()
```

Now I want to specify my model. I decided to use a Gradient Boosting model.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

# split dataset into test and train
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25, random_state = 0)

gbc = GradientBoostingClassifier(max_depth = 4, random_state = 0)

gbc.fit(xtrain, ytrain)
pred = gbc.predict(xtest)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay 
import matplotlib.pyplot as plt

cm = confusion_matrix(ytest, pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm) 
disp.plot() 
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(ytest, pred)
precision = precision_score(ytest, pred)
recall = recall_score(ytest, pred)
f1 = f1_score(ytest, pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-Score: {f1}')
```

The model achieved a strong accuracy score, however performed poorly on the minority class and this is reflected in the low scores for all of the other metrics. To achieve better model performance I want to repeat my analysis but with resampling.

## Reapply GBC with resampling

Using random oversampling method. We only oversample the training set, otherwise we get information leakage whereby the observations from the test set are used to train the model.

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0) 

xtrain_s, ytrain_s = ros.fit_resample(xtrain, ytrain)
```

```python
# check that dataset is now balanced
ytrain.value_counts(), ytrain_s.value_counts()
```

```python
# train the model
gbc.fit(xtrain_s, ytrain_s)
```

```python
pred = gbc.predict(xtest)
```

```python
cm = confusion_matrix(ytest, pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm) 
disp.plot() 
plt.show()
```

```python
accuracy = accuracy_score(ytest, pred)
precision = precision_score(ytest, pred)
recall = recall_score(ytest, pred)
f1 = f1_score(ytest, pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-Score: {f1}')
```

Using the resampled training data, the accuracy of the model has dropped to 81% however the F1 score has increased because the model now performs well on both classes.

## Cross Validation

I have seen that resampling the data can lead to improved results. My next step will be to do this in combination with cross validation (in this case I am using stratified K-folds cross validation). Cross validation means that all of the data is used to train the model and ensures more reliable results since multiple subsets are used for both training and testing the model.

It is important that we create a pipeline whereby we can chain multiple stages of the process together. In this instance, we chain the random oversampling and the classification model together so that the oversampling is done within each fold; this prevents information leakage, whereby data from the test set is inadvertedly used to train the model.

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler

# Create a pipeline that applies oversampling within each fold
pipeline = Pipeline([('oversampler', ros), ('classifier', gbc)])

# Define the cross-validation strategy
kf = StratifiedKFold(n_splits=4, shuffle=True, random_state=0)

# Perform cross-validation
scores = cross_val_score(pipeline, X, y, cv=kf, scoring='f1')

# Print the cross-validation scores
print("Cross-validation F1 scores:", scores)
print("Average cross-validation F1 score:", scores.mean())
```

```python
acc = cross_val_score(pipeline, X, y, cv=kf, scoring='accuracy')

# Print the cross-validation scores
print("Cross-validation accuracy scores:", acc)
print("Average cross-validation accuracy score:", acc.mean())
```

```python
from sklearn.model_selection import cross_val_predict

# Get cross-validated predictions 
y_pred = cross_val_predict(pipeline, X, y, cv=kf) 
# Compute the confusion matrix 
cm = confusion_matrix(y, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()
```

# Hyperparameter Tuning

Next step is to use grid search to tune the model hyperparameters in order to optimise performance. The hyperparameters that I am trying to optimise are max depth, n estimators and learning rate.

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid
param_grid = {
    'classifier__max_depth': [3, 4, 5],
    'classifier__n_estimators': [100, 200, 300],
    'classifier__learning_rate': [0.1, 0.2, 0.3]
}

# Combine Grid Search and Cross-Validation
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=kf, scoring='f1', n_jobs=1)

# Fit Grid Search to the data
grid_search.fit(X, y)

# Print the best hyperparameters and the best F1 score
print(f"Best hyperparameters: {grid_search.best_params_}")
print(f"Best F1 score: {grid_search.best_score_}")
```

```python
best_model = grid_search.best_estimator_

# Get cross-validated predictions 
y_pred = cross_val_predict(best_model, X, y, cv=kf) 

# Compute the confusion matrix 
cm = confusion_matrix(y, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot()
plt.show()
```

```python
f1 = cross_val_score(pipeline, X, y, cv=kf, scoring='f1')
acc = cross_val_score(pipeline, X, y, cv=kf, scoring='accuracy')

print(acc)
```

# Feature Importance

```python
classifier = best_model.named_steps['classifier']

# Extract feature importances
feature_importances = classifier.feature_importances_

# Assuming `X` contains your feature names
features = X.columns

# Create a DataFrame for better plotting
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})

# Sort the DataFrame by importance
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Sort the DataFrame by importance and select top 10 features 
top_10_importance_df = importance_df.sort_values(by='Importance', ascending=False).head(10)

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.barh(top_10_importance_df['Feature'], top_10_importance_df['Importance'], color='skyblue')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importances from Gradient Boosting Classifier')
plt.gca().invert_yaxis()  # To display the highest importance at the top
plt.show()
```

The features ranked in order of importance are:
1. Whether the person found their date attractive
2. Whether their date found that person attractive
3. Whether their date found that person funny
4. Whether the person found their date funny
5. The probability that the person thought their date liked them
6. Whether their date thought that they had shared interests
7. The expected number of matches that the person thought they would get across all of their dates
8. Whether the person thought that they had shared interests
9. How important that their date thought intelligence was in a partner
10. Correlation between how the person and the date scored their interests

In summary, the three most important factors in whether a couple will match or not appear to be whether they find each other attractive, whether they find each other funny and whether they have shared interests.

# Conclusions

I used a Gradient Boosting classification model to try to predict whether couples in a speed dating experiment would match. My dataset was very imbalanced (as most couples don't match) and, without resampling, I found that my model was very accurate but biased towards the majority class. When I applied random oversampling, the accuracy of my model fell by around 5% but my F1 score increased.

I subsequently used grid search to optimise my hyperparamter and found that this increased my F1 score by about 3%. My best model was approximately 79% accurate. The model is highly accurate for the majority class (non-matches) and around 60% accurate for the minority class (matches).

I conducted feature engineering and found that the most important factors in whether a couple will match are:
1. Whether they find each other attractive
2. Whether they find each other funny
3. Whether they have shared interests

I believe that I could have achieved better results with additional feature engineering, however the model is able to classify matches and non-matches with reasonable success. Additionally, a different classification model could have been more successful so if I had more time I would try multiple models and compare success rather than just selecting one.
